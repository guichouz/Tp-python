```python

from tkinter import *
from math import *

def passage(x,y, a, b):
    return (a+x, b-y)

def produit(k, n, p):
    return k*n % p

def segment(cnv, k, n, p, base, a, b):
    # base : les coordonnées des points du cercle
    # on trace le produit prod=k*n
    # k -> A
    # prod -> B
    prod=produit(k, n, p)
    xA, yA=base[k]
    xB, yB=base[prod]    
    A=passage(xA, yA, a, b)
    B=passage(xB, yB, a, b)
    cnv.create_line(A, B)
    
def dot(cnv, C, R=6, color='red'):
    xC, yC=C
    return cnv.create_oval(xC-R,yC-R,xC+R, yC+R, fill=color, outline=color)
  
def draw_base(cnv, p, base, a, b):
    p=len(base)
    # Les sommets sur le cercle
    for k in range(p):
        x, y=base[k]
        X, Y=passage(x, y, a, b)
        dot(cnv, (X, Y), R=2, color='black')
        
def table(value):
    global N
    N=int(value)
    show(N, P)

def point(value):
    global P
    P=int(value)
    show(N, P)
    
def show(n,p):
    cnv.delete("all")
    t=2*pi/p
    base=[(R*cos(2*k*pi/p), R*sin(2*k*pi/p)) for k in range(p)]
    draw_base(cnv, p, base, a, b)
    # Tracé de la table
    for k in range(p):
        segment(cnv, k, n, p, base, a, b)  
    #master.mainloop()
    
p=100 # nombre de points sur le cercle
R=200
n=5 # Table de n    

N=P=2
    
master=Tk()
a=b=1.2*R
cnv=Canvas(master, width=2*a, height=2*b, bg='ivory')
cnv.pack(side="left")

    
table_curseur= Scale(master, label="Table de ...",
                     orient = "horizontal",
                     font="Arial 15 bold",
                    command = table, 
                    from_=2, to=50)   
table_curseur.pack(pady=15)
    
point_curseur= Scale(master, label="nombre de points",
                    orient= "horizontal",
                     font="Arial 15 bold",
                     command= point,
                     from_=2, to=200,
                    length=250)
point_curseur.pack(pady=5)
    
    
    
master.mainloop()


```

```python
from tkinter import *

TOP=20
LEFT=80
RIGHT=80
BASE_H=TOP

SEP=8
N=6
P=7
RAD=20
DIAM=2*RAD

PERIOD=2*SEP+DIAM


BOARD_W=P*PERIOD
BOARD_H=N*PERIOD

WIDTH=LEFT+RIGHT+P*PERIOD
HEIGHT=BASE_H+N*PERIOD+TOP

def dot(cnv, C, R=6, color='red'):
    xC, yC=C
    return cnv.create_oval(xC-R,yC-R,xC+R, yC+R, fill=color, outline=color)

def play_token(cnv, line, col, color):
    x0=LEFT+0.5*PERIOD
    y0=TOP+0.5*PERIOD+(N-1)*PERIOD
    xC=x0+col*PERIOD
    yC=y0 - line*PERIOD 
    dot(cnv, (xC,yC), RAD, color=color)

def fill_board(cnv):
    x=x0=LEFT+0.5*PERIOD
    y=y0=TOP+0.5*PERIOD

    for i in range(P):
        for j in range(N):
            dot(cnv, (x,y), RAD, color='white')
            y+=PERIOD

        x+=PERIOD
        y=y0

    x=x0+3*PERIOD
    y=y0+(N-1)*PERIOD

def show():
    root = Tk()
    cnv = Canvas(root, width=WIDTH, height=HEIGHT, background="ivory")
    cnv.pack()

    base=cnv.create_rectangle((0, HEIGHT), (WIDTH, HEIGHT-BASE_H), fill="blue", outline='')
    board=cnv.create_rectangle((LEFT, TOP), (WIDTH-RIGHT, HEIGHT-BASE_H), fill="blue", outline='')
    fill_board(cnv)
    

    root.mainloop()
        

show()
```
